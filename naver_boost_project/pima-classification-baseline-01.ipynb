### scikit-learn: python의 대표적인 머신러닝 라이브러리
주로 supervised learning(분류,회귀)을 다룰 것임

```python
# Decision Tree로 분류하기

from sklearn import tree
X = [[0,0],[1,1]]
Y = [0,1]
```

```python
clf=tree.DecisionTreeClassifier()
clf
```

```python
clf=clf.fit(X,Y)
clf
```

```python
clf.predict([[2.,2.,]])
```

```python
clf.predict_proba([[2.,2.,]])
```

```python
# Iris data로 Decision Tree Classification하기

from sklearn.datasets import load_iris
from sklearn import tree
x,y=load_iris(return_X_y=True)
x,y
```

```python
clf=tree.DecisionTreeClassifier()
clf
```

```python
clf=clf.fit(x,y)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(20,10))
t=tree.plot_tree(clf.fit(x,y),filled=True)
t
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
df=pd.read_csv('./diabetes.csv')
df.shape
```

```python
df.head()
```

```python
# train-test split

split_count=int(df.shape[0]*0.8)
split_count
```

```python
train=df[:split_count].copy()
test=df[split_count:].copy()
```

```python
print(train.shape,test.shape)
```

```python
feature_names=train.columns[:-1].tolist()
feature_names
```

```python
label_name=train.columns[-1]
label_name
```

```python
train[feature_names]
# -> outcome column이 빠짐
```

```python
x_train=train[feature_names]
print(x_train.shape)
x_train.head()
```

```python
train[label_name]
```

```python
y_train=train[label_name]
print(y_train.shape)
y_train.head()
```

```python
x_test=test[feature_names]
print(x_test.shape)
x_test.head()
```

```python
y_test=test[label_name]
print(y_test.shape)
```

# -----

```python
from sklearn.tree import DecisionTreeClassifier

model=DecisionTreeClassifier()
model
```

```python
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
```

```python
y_pred
```

```python
y_test
```

```python
# 정확도구하기
(y_pred==y_test).mean()
```
