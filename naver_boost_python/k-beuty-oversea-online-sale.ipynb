```python
# 데이터 전처리: tidy data만들기(melt함수 이용)
```

![image](https://user-images.githubusercontent.com/47767202/91116688-dbe01e80-e6c7-11ea-8ff2-01e62958371d.png)

```python
df=df_raw.melt(id_vars=['국가(대륙)별','상품군별','판매유형별'],var_name='기간', value_name='백만원')
df
```

```python
df.info()
```

# -----

```python
# 기간에서 연도 분리하기
df['기간']
```

```python
df['기간'].map(lambda x:int(x.split()[0]))
```

```python
df['연도']=df['기간'].map(lambda x:int(x.split()[0]))
df.head()
```

```python
df['기간'].map(lambda x:int(x.split()[1].split('/')[0]))
```

```python
df['분기']=df['기간'].map(lambda x:int(x.split()[1].split('/')[0]))
df.head()
```

```python
# 금액데이터를 수치로 변환하기 위해 타입변경
df['백만원']
```

```python
df['백만원'].replace('-',pd.np.nan)
```

```python
df['백만원'].replace('-',pd.np.nan).astype(float)
```

```python
df['백만원']=df['백만원'].replace('-',pd.np.nan).astype(float)
df.head()
```

```python
# 필요없는 컬럼 제거

(df['국가(대륙)별']=='합계') | (df['상품군별']=='합계')
```

```python
df.info()
```

```python
df=df[(df['국가(대륙)별']!='합계') & (df['상품군별']!='합계')].copy()
```

```python
df.info()
```

```python
# 결측치 확인

df.isnull().sum()
```

```python
df_sub=df_total[~df_total['상품군별'].isin(['화장품','의류 및 패션관련 상품'])].copy()
```

```python
sns.relplot(data=df_sub,x='연도',y='백만원',hue='상품군별',kind='line',
            col='상품군별',col_wrap=4)
```
